Firstly I load the dataset and read the data in order to operate the data. 

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/postoperative-patient-data/post-operative.data'
raw_data = urlopen(url)
```

```python
dataset = np.genfromtxt(raw_data, delimiter=",", dtype=str)
```

I get the shape of the dataset and pass the dataset to X, and copy X to modified_X, then I use the nan to replace the missing data

```python
dataset.shape
```

```python
X = dataset[:,0:89]
```

```python
modified_X = np.copy(X)
modified_X[modified_X=='?']=np.nan
```

```python
X
```

```python
modified_X
```

I transform the array to the dataframe because the dataframe type is easier to operate and more availabe to display

```python
import pandas as pd
df=pd.DataFrame(modified_X,
               columns=['L-CORE', 'L-SURF', 'L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','decision ADM-DECS'])
df
```

I change the string level to numeric representation because numberic varibles are more available to handle missing data 
and display by using coordinate system. I convert the columns with the same level of criteria in a few steps. 

```python
size_mapping = {'low': 1,'mid': 2,'high': 3}
```

```python
df['L-CORE'] = df['L-CORE'].map(size_mapping)
df['L-SURF'] = df['L-SURF'].map(size_mapping)
df['L-BP'] = df['L-BP'].map(size_mapping)
df
```

```python
size_mapping = {'poor': 1,'fair': 2,'good': 3,'excellent': 4 }
```

```python
df['L-O2'] = df['L-O2'].map(size_mapping)
df
```

```python
size_mapping = {'unstable': 1,'mod-stable': 2,'stable': 3}
```

```python
df['SURF-STBL'] = df['SURF-STBL'].map(size_mapping)
df['CORE-STBL'] = df['CORE-STBL'].map(size_mapping)
df['BP-STBL'] = df['BP-STBL'].map(size_mapping)
df
```

```python
size_mapping = {'S': 1,'A': 2,'I': 3}
```

```python
df['decision ADM-DECS'] = df['decision ADM-DECS'].map(size_mapping)
df
```

I use the mean number of the column to replace the missing date in the column

```python
from sklearn.preprocessing import Imputer

imr = Imputer(missing_values='NaN', strategy='mean', axis=0)
imr = imr.fit(df)
imputed_data = imr.transform(df.values)
imputed_data
```

After caculating the missing data, I convert the array to dataframe again.

```python
df=pd.DataFrame(imputed_data,columns=['L-CORE', 'L-SURF', 'L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','decision ADM-DECS'])
df
```
